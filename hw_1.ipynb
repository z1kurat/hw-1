```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# параметры для генерации выборки стандартного нормального распределения
np.random.seed(0) # Для воспроизводимости результатов
mu, sigma = 0, 1 # Параметры нормального распределения
```

```python
def ecdf(sample, x):
    """Построение эмпирическая функция распределения по определению"""
    sorted_sample = np.sort(sample)
    indices = np.searchsorted(sorted_sample, x, side='right')
    return indices / len(sample)
```

```python
def plot_ecdf(sample, n, title):
    """Построение графиков эмпирической функции распределения и истинной функции распределения"""
    x = np.linspace(min(sample), max(sample), 100)
    y_ecdf = ecdf(sample, x)
    y_true = norm.cdf(x, mu, sigma) # Истинная функция распределения

    plt.plot(x, y_ecdf, 'b-', label='Эмпирическая функция распределения')
    plt.plot(x, y_true, 'r-', label='Истинная функция распределения')
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('F(x)')
    plt.legend()
    plt.show()
```

```python
# Генерация выборок разного размера
n_values = [10, 1000, 10000, 100000]
samples = [np.random.normal(mu, sigma, n) for n in n_values]

# Построение графиков, которые демонстрируют приближение эмпирической функции
# распределения к исходной функциии распределения с размером выборки.
for n, sample in zip(n_values, samples):
    plot_ecdf(sample, n, f'Эмпирическая функция распределения и истинная функция распределения для n={n}')
```

```python
# Генерация выборок разного размера
n_values = [10, 100, 250, 500, 750, 1000, 2500, 5000, 7500, 10000, 25000, 50000, 75000, 100000]
samples = [np.random.normal(mu, sigma, n) for n in n_values]
```

```python
def plot_mean(samples, n_values, title):
    """Построение графика среднего значения выборки."""
    means = [np.mean(sample) for sample in samples]
    plt.plot(n_values, means, 'bo-', label='Среднее значение выборки')
    plt.axhline(y=mu, color='r', linestyle='-', label='Ожидаемое значение')
    plt.title(title)
    plt.xlabel('Размер выборки (n)')
    plt.ylabel('Среднее значение')
    plt.legend()
    plt.show()
```

```python
# Пусть X1, ..., Xn — выборка из нормального распределения с параметрами a и σ
# Тогда имеет место оценка: 1 / n * \sum_i X_i = a
# В нашем случаем - a = 0
plot_mean(samples, n_values, 'Сходимость среднего значения выборки')
```
